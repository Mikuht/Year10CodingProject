```python
#Setup

import json
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'figure.autolayout': True})
plt.style.use('fivethirtyeight')
```

```python
#Function Def

def sortdic(diction):
    diction = dict(sorted(diction.items(), key=lambda item: item[1],reverse=True))
    return diction

def percentage(x, pos):
    s = '{:1.0f}%'.format(x)
    return s

def get_key(val, my_dict):
    for key, value in my_dict.items():
         if val == value:
             return key
 
    return "key doesn't exist"
```

```python
#Load json File
#Create counting list format

characters = json.load(open("characters.json"))
abyss = json.load(open("v2.0-star36.json"))
charList = {}
for i in characters:
    charList[i] = 0
# print(charList)
```

```python
#Count Ownership data from json
#Count Usage data from json
#Calculate Average Level of Each Character
#Calculate Average Constellation of Each Character
#Count Team Usage data from json

level9team = {"f9_1_a" : {}, "f9_1_b" : {}, "f9_2_a" : {},"f9_2_b" : {},"f9_3_a" : {},"f9_3_b" : {}}
level10team = {"f10_1_a" : {}, "f10_1_b" : {}, "f10_2_a" : {},"f10_2_b" : {},"f10_3_a" : {},"f10_3_b" : {}}
level11team = {"f11_1_a" : {}, "f11_1_b" : {}, "f11_2_a" : {},"f11_2_b" : {},"f11_3_a" : {},"f11_3_b" : {}}
level12team = {"f12_1_a" : {}, "f12_1_b" : {}, "f12_2_a" : {},"f12_2_b" : {},"f12_3_a" : {},"f12_3_b" : {}}
teams = {}
own = charList.copy()
averageLevel = charList.copy()
averageConstellation = charList.copy()
level9 = {"f9_1_a" : charList.copy(), "f9_1_b" : charList.copy(), "f9_2_a" : charList.copy(),"f9_2_b" : charList.copy(),"f9_3_a" : charList.copy(),"f9_3_b" : charList.copy()}
level10 = {"f10_1_a" : charList.copy(), "f10_1_b" : charList.copy(), "f10_2_a" : charList.copy(),"f10_2_b" : charList.copy(),"f10_3_a" : charList.copy(),"f10_3_b" : charList.copy()}
level11 = {"f11_1_a" : charList.copy(), "f11_1_b" : charList.copy(), "f11_2_a" : charList.copy(),"f11_2_b" : charList.copy(),"f11_3_a" : charList.copy(),"f11_3_b" : charList.copy()}
level12 = {"f12_1_a" : charList.copy(), "f12_1_b" : charList.copy(), "f12_2_a" : charList.copy(),"f12_2_b" : charList.copy(),"f12_3_a" : charList.copy(),"f12_3_b" : charList.copy()}
for i in abyss:
    for j in i:
        if j == "roles":
            for x in i[j]:
                own[str(x)] += 1
        if j == "roles_detail":
            for x in i[j]:
                averageLevel[str(x)] += i[j][x]["level"]
                averageConstellation[str(x)] += i[j][x]["constellations"]
        if j in level9:
            temp = i[j]
            if str(temp) in level9team[j]:
                level9team[j][str(temp)] += 1
            else:
                if not temp in teams.values():
                    teams[str(temp)] = temp
                level9team[j][str(temp)] = 1
            for x in i[j]:
                level9[j][str(x)] += 1
        if j in level10:
            temp = i[j]
            if str(temp) in level10team[j]:
                level10team[j][str(temp)] += 1
            else:
                if not temp in teams.values():
                    teams[str(temp)] = temp
                level10team[j][str(temp)] = 1
            for x in i[j]:
                level10[j][str(x)] += 1
        if j in level11:
            temp = i[j]
            if str(temp) in level11team[j]:
                level11team[j][str(temp)] += 1
            else:
                if not temp in teams.values():
                    teams[str(temp)] = temp
                level11team[j][str(temp)] = 1
            for x in i[j]:
                level11[j][str(x)] += 1
        if j in level12:
            temp = i[j]
            if str(temp) in level12team[j]:
                level12team[j][str(temp)] += 1
            else:
                if not temp in teams.values():
                    teams[str(temp)] = temp
                level12team[j][str(temp)] = 1
            for x in i[j]:
                level12[j][str(x)] += 1
for i in averageConstellation:
    averageConstellation[i] /= 1000.00
    # print(characters[i] + ":" + str(averageConstellation[i]))
for i in averageLevel:
    averageLevel[i] /= 1000.00
    # print(characters[i] + ":" + str(averageLevel[i]))
for i in level9team:
    level9team[i] = sortdic(level9team[i])
    # for j in level9team[i]:
    #     print(j + ": " + str(level9team[i][str(j)]))
    # print()
for i in level10team:
    level10team[i] = sortdic(level10team[i])
for i in level11team:
    level11team[i] = sortdic(level11team[i])
for i in level12team:
    level12team[i] = sortdic(level12team[i])
```

```python
# Transform Ownership count into percentage
# Showing Counting Datas with character names

# print("ownership")
ownRate = charList.copy()
for a in own:
    # print(characters[a] + ":" + str(own[a]))
    ownRate[a] = own[a] / 1000.00
# print()
# # ownRate = sortdic(ownRate)
# for a in ownRate:
#     print(characters[a] + ":" + str(ownRate[a]))
# for a in level9:
#     print()
#     print(a)
#     for b in level9[a]:
#         print(characters[b] + ":" + str(level9[a][b]))
# for a in level10:
#     print()
#     print(a)
#     for b in level10[a]:
#         print(characters[b] + ":" + str(level10[a][b]))
# for a in level11:
#     print()
#     print(a)
#     for b in level11[a]:
#         print(characters[b] + ":" + str(level11[a][b]))
# for a in level12:
#     print()
#     print(a)
#     for b in level12[a]:
#         print(characters[b] + ":" + str(level12[a][b]))
```

```python
#Find the level that each character was most used in
#Create list of using count of characters in each level

mostUsed = {}
useCount = {}
for i in characters:
    mostUsed[i] = ""
    useCount[i] = [0,0,0,0]
for c in characters:
    for i in level9:
        useCount[c][0] += level9[i][c]
    for i in level10:
        useCount[c][1] += level10[i][c]
    for i in level11:
        useCount[c][2] += level11[i][c]
    for i in level12:
        useCount[c][3] += level12[i][c]
    # print(characters[c])
    # for i in useCount[c]:
    #     print(i)
    match np.argmax(useCount[c]):
        case 0:
            mostUsed[c] = "level 9"
        case 1:
            mostUsed[c] = "level 10"
        case 2:
            mostUsed[c] = "level 11"
        case 3:
            mostUsed[c] = "level 12"

# for i in mostUsed:
#     print(characters[i] + ": "+  str(mostUsed[i]))
```

```python
#Transform Usage count into percentage

usage = {"f9_1_a" : charList.copy(), "f9_1_b" : charList.copy(), "f9_2_a" : charList.copy(),"f9_2_b" : charList.copy(),"f9_3_a" : charList.copy(),"f9_3_b" : charList.copy(), "f10_1_a" : charList.copy(), "f10_1_b" : charList.copy(), "f10_2_a" : charList.copy(),"f10_2_b" : charList.copy(),"f10_3_a" : charList.copy(),"f10_3_b" : charList.copy(), "f11_1_a" : charList.copy(), "f11_1_b" : charList.copy(), "f11_2_a" : charList.copy(),"f11_2_b" : charList.copy(),"f11_3_a" : charList.copy(),"f11_3_b" : charList.copy(), "f12_1_a" : charList.copy(), "f12_1_b" : charList.copy(), "f12_2_a" : charList.copy(),"f12_2_b" : charList.copy(),"f12_3_a" : charList.copy(),"f12_3_b" : charList.copy()}
for a in level9:
    # print()
    # print(a)
    for b in level9[a]:
        usage[a][b] = level9[a][b] / own[b]
        # print(characters[b] + ":" + str(usage[a][b]))
for a in level10:
    # print()
    # print(a)
    for b in level10[a]:
        usage[a][b] = level10[a][b] / own[b]
        # print(characters[b] + ":" + str(usage[a][b]))
for a in level11:
    # print()
    # print(a)
    for b in level11[a]:
        usage[a][b] = level11[a][b] / own[b]
        # print(characters[b] + ":" + str(usage[a][b]))
for a in level12:
    # print()
    # print(a)
    for b in level12[a]:
        usage[a][b] = level12[a][b] / own[b]
        # print(characters[b] + ":" + str(usage[a][b]))
```

```python
#Calculate total (average) Usage

totalUsage = charList.copy()
for a in usage:
    for b in usage[a]:
        totalUsage[b] += usage[a][b]

for i in totalUsage:
    totalUsage[i] /= 12
    # print(characters[i] + ":" + str(totalUsage[i]))

# totalUsage = sortdic(totalUsage)

# for i in totalUsage:
#     print(characters[i] + ":" + str(totalUsage[i]))
```

```python
#Output Chart - Rate of Owning and Using

sortOwn = {}
for i in sortdic(ownRate):
    sortOwn[i] = ownRate[i] * 100
sortUse = {}
for i in sortOwn:
    sortUse[i] = totalUsage[i] * 100

useName = []
for i in sortUse:
    useName.append(characters[i])
ownName = []
for i in sortOwn:
    ownName.append(characters[i])

y=np.arange(len(ownName))
width = 0.4
fig, charG = plt.subplots(figsize=(12,20))
charG.set(xlim=[0,100.3], ylabel = "Characters", title = "Rate of Owning & Using Characters", yticks = y, yticklabels = ownName)
ownBar = charG.barh(y - width/2, list(sortOwn.values()), width, label = "owning rate")
useBar = charG.barh(y + width/2, list(sortUse.values()), width, label = "using rate")
charG.legend()
charG.bar_label(ownBar, padding = 3)
charG.bar_label(useBar, padding = 3)
plt.suptitle("(sorted by owning rate)")

charG.xaxis.set_major_formatter(percentage)
plt.show()
```
